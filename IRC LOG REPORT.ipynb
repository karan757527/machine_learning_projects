```python
# importing required modules
import pandas as pd
import numpy as np
import re
from datetime import datetime
from datetime import timedelta
```

```python
#reading hackers.log file in logs variable
with open("Data\\hackers.log",'r',encoding="latin1") as fp:
    logs=fp.read()
    fp.close()
```

```python
# getting data date wise
exp=r"^--- Day.*[\d]{4}"
day=[]
time=[]
data={}
x=0
for day in re.finditer(exp,logs,flags=re.M):
    z=" ".join(day[0].split()[4:])
    y=str(datetime.strptime(z,"%b %d %Y").date()-timedelta(days=1))
    data[y]=logs[x:day.span()[0]]
    x=day.span()[1]
print(z)
data[str(datetime.strptime(z,"%b %d %Y").date())]=logs[x:]
```

```python
# getting lists
datec=[]
timec=[]
msgc=[]
x=0
for i in data:
    x+=1
    for j in data[i].split("\n")[1:]:
        #print(j)
        if j and not (j[0]=="-" or j[0]==" " or j[0]=="<"):
            datec.append(i)
            timec.append(j.split(" ")[0])
            msgc.append(" ".join(j.split()[1:]))
```

```python
# making dataframe
df=pd.DataFrame(data={
   "Date":datec,
    "Time":timec,
    "Message":msgc 
    
},columns=[
    "Date",
    "Time",
    "Message"
])
```

```python
# extract name from msg
def user_name(x):
    if not x:
        return "NA"
    elif x[0]=="-":
        if len(x.split(" is now known as"))==1:
            if len(x.split(" [")[0])<5:
                return " [".join(x.split(" [")[:2])
            
            return x.split(" [")[0]
        else:
            return x.split(" is")[0]
    else:
        return x.split(">")[0]+">"
df["User"]=df["Message"].apply(user_name)
```
